```python
import requests
from bs4 import BeautifulSoup
```

```python
site = requests.get('https://newmapsplus.as.uky.edu/')
```

```python
nmp_raw = BeautifulSoup((site.text), "html.parser")
```

```python
for link in nmp_raw.findAll('a'):
    print(link.get('href'))
```

```python
print(nmp_raw.getText())
```
